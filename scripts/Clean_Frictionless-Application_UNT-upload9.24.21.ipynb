```python
#import 'describe' function from frictionless
from frictionless import describe
#import 'pretty printer' module for more readable dictionary data
import pprint

pp = pprint.PrettyPrinter(depth=4)
```

```python
#use 'describe' to generate a 'Frictionless table schema' for each dataset

NP_Titles_schema = describe ('G:\\My Drive\\Library\\Projects\\LEADING\\revisions202108\\Texas_NewspaperTitleList_final.csv')

TX_IDs_schema = describe ('G:\\My Drive\\Library\\Projects\\LEADING\\revisions202108\\TexasCounty_IDList.csv')
```

```python
#print description
pp.pprint(NP_Titles_schema)
pp.pprint(TX_IDs_schema)
```

```python
#adding field descriptors to table schema 
###NP_Titles_schema
NP_Titles_schema.schema.get_field("title").title = "Newspaper Title"
NP_Titles_schema.schema.get_field("title").description = "The title of the newspaper as documented by Chronicaling America https://chroniclingamerica.loc.gov/"

NP_Titles_schema.schema.get_field("lccn").title = "lccn"
NP_Titles_schema.schema.get_field("lccn").description = "Library of Congress Control Number - LCCN"

NP_Titles_schema.schema.get_field("start_year").title = "Start Year"
NP_Titles_schema.schema.get_field("start_year").description = "Year of first publication from this newspaper"

NP_Titles_schema.schema.get_field("end_year").title = "End Year"
NP_Titles_schema.schema.get_field("end_year").description = "Year of last publication from this newspaper"

NP_Titles_schema.schema.get_field("frequency").title = "Frequency - Original"
NP_Titles_schema.schema.get_field("frequency").description = "Frequency of publication as originally documented by Chronicaling America"

NP_Titles_schema.schema.get_field("frequency_normalized").title = "Frequency - Normalized"
NP_Titles_schema.schema.get_field("frequency_normalized").description = "Frequency normalized manually to include primary headings (e.g., daily, weekly)"

NP_Titles_schema.schema.get_field("language").title = "Language of Publications"
NP_Titles_schema.schema.get_field("language").description = "Languages that this newspaper publishes in"

NP_Titles_schema.schema.get_field("county").title = "Distribution Counties"
NP_Titles_schema.schema.get_field("county").description = "The Texas county or counties in which this newspaper distributes publications - extracted via excel from the original 'place' variable"

NP_Titles_schema.schema.get_field("place").title = "Place - Original"
NP_Titles_schema.schema.get_field("place").description = "Place of distribution in Texas as originally documented by Chronicaling America - Non-Texas locations removed manually"

NP_Titles_schema.schema.get_field("place_of_publication").title = "Place of Publication"
NP_Titles_schema.schema.get_field("place_of_publication").description = "Place(s) where newspaper is published as originally documented by Chronicaling America"

pp.pprint(NP_Titles_schema)
```

```python
####TX_IDs_schema 
TX_IDs_schema.schema.get_field("County Name").title = "County Name"
TX_IDs_schema.schema.get_field("County Name").description = "Name of all 254 Counties in Texas"


TX_IDs_schema.schema.get_field("FIPS Code - State").title = "FIPS Code - State"
TX_IDs_schema.schema.get_field("FIPS Code - State").description = "Federal Information Processing Standards (FIPS) code for Texas state"

TX_IDs_schema.schema.get_field("FIPS Code - County").title = "FIPS Code - County"
TX_IDs_schema.schema.get_field("FIPS Code - State").description = "Federal Information Processing Standards (FIPS) code for each county in Texas"

TX_IDs_schema.schema.get_field("geonames ID").title = "geonames"
TX_IDs_schema.schema.get_field("geonames ID").description = "Geonames Identifier as documented by https://www.geonames.org/"

TX_IDs_schema.schema.get_field("wikidata ID").title = "wikidata"
TX_IDs_schema.schema.get_field("wikidata ID").description = "Wikidata Identifier as documented by https://www.wikidata.org/wiki/Wikidata:Identifiers"

TX_IDs_schema.schema.get_field("UNTL County String").title = "UNTL County String"
TX_IDs_schema.schema.get_field("UNTL County String").description = "University of North Texas Library County String Identifier; formmated as '[County Name], TX'"

TX_IDs_schema.schema.get_field("UNTL Place Name String").title = "UNTL Place Name String"
TX_IDs_schema.schema.get_field("UNTL Place Name String").description = "University of North Texas Library Place Name String Identifier; formatted as 'United States - Texas - [County Name]'"

pp.pprint(TX_IDs_schema)
```

```python
#adding constraint field descriptors to the tables

NP_Titles_schema.schema.get_field("frequency_normalized").constraints["enum"] = ["Daily", "Semiweekly", "Weekly", "Semimonthly", "Monthly", "Bimonthly", "Quarterly", "Annual", "Irregular", "Missing"]
pp.pprint(NP_Titles_schema)

TX_IDs_schema.schema.get_field("FIPS Code - State").constraints["enum"] = ["48"]
pp.pprint(TX_IDs_schema)
```

```python
#handling missing values
NP_Titles_schema.schema.missing_values = ["","1000","9999","Missing"]
pp.pprint(NP_Titles_schema)

TX_IDs_schema.schema.missing_values = [""]
pp.pprint(TX_IDs_schema)
```

```python
#adding a table description
NP_Titles_schema.description = "The titles, lccn, years of activity, frequency and place of distribution, and place of publication of texas newspapers as documented by Chronicaling America"
pp.pprint(NP_Titles_schema)

TX_IDs_schema.description = "The FIPS code, geonames, and wikidata identifiers, as well as the University of North Texas Library's (UNTL) County and Place name string formats for all Texas Counties"
pp.pprint(TX_IDs_schema)
```

```python
#adding a primary key to TX_IDs
TX_IDs_schema.schema.primary_key = "County Names"
pp.pprint(TX_IDs_schema)
```

```python
#applying foreign key to NP_Titles

f_keys = []
f_keys.append({
  "fields": "county",
  "reference": {
      "resource": "TexasCounty_IDList.csv",        
      "fields": "County Names"
  }            
})

NP_Titles_schema.schema.foreign_keys = f_keys
pp.pprint(NP_Titles_schema)
```
